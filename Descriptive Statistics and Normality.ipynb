```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

span = np.random.gamma(70,1, 100)

plt.hist(span)

plt.axvline(span.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(span.mean() + span.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(span.mean()-span.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

The distribution is mostly symmetrical and clustered in the center, and so we can say that the mean reflects the central tendency of the variable and the standard deviation roughly describes the average variance.

```python
totallynormal = np.random.normal(5,0.5,10000)
absolutelynormal = np.random.normal(10,1,10000)
completelynormal = totallynormal + absolutelynormal
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(totallynormal)

plt.axvline(totallynormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(totallynormal.mean() + totallynormal.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(totallynormal.mean()-totallynormal.std(), color='b', linestyle='dashed', linewidth=2) 

plt.subplot(1, 3, 2)
plt.hist(absolutelynormal)

plt.axvline(absolutelynormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(absolutelynormal.mean() + absolutelynormal.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(absolutelynormal.mean()-absolutelynormal.std(), color='b', linestyle='dashed', linewidth=2) 

plt.subplot(1, 3, 3)
plt.hist(completelynormal)

plt.axvline(completelynormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(completelynormal.mean() + completelynormal.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(completelynormal.mean()-completelynormal.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```
