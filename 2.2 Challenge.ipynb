```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.boxplot(df['math score'])
plt.ylabel('Percent Correct')
plt.title('Math Scores')

plt.subplot(1, 3, 2)
plt.boxplot(df['reading score'])
plt.ylabel('Percent Correct')
plt.title('Reading Scores')

plt.subplot(1, 3, 3)
plt.boxplot(df['writing score'])
plt.ylabel('Percent Correct')
plt.title('Writing Scores')

plt.tight_layout()
plt.show()
```

We can see that in all three plots, the students scores on each test, the inter-quartile range begins (with the 25th percentile) at or near a score of 60, and ends (with the 75th percentile) just below 80. This can be interpreted as more than 50% of the students achieving a letter grade of 'D' or 'C'. It can be seen that the median grade for all 3 tests is in the 'D' range meaning more than half of students earned a 'D' or below.

```python
masters_df = df[df['parental level of education'] == "master's degree"]
bachelors_df = df[df['parental level of education'] == "bachelor's degree"]

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.hist(bachelors_df['math score'], density = True)
plt.ylabel('Density')
plt.xlabel('Math Score')
plt.title("Bachelors' Children's Math Scores")

plt.subplot(1,2,2)
plt.hist(masters_df['math score'], density=True)
plt.ylabel('Density')
plt.xlabel('Math Score')
plt.title("Masters' Children's Math Scores")


plt.tight_layout()
plt.show()
```

Two data sets,  the math scores of Students' whose parents have a Bachelor's degree and the math scores of students' whose parents have a Master's degree, were compared by two histograms. These two graph show that while both data sets have similar means, and both are skewed two the left, the Master's degree scores tended to be more uniform in distribution, and more frequently near the extreme values.

```python
male_df = df[df['gender'] == "male"]
female_df = df[df['gender'] == "female"]

male = plt.scatter(x = male_df['math score'],y = male_df['reading score'], color = 'blue')
female = plt.scatter(x = female_df['math score'],y = female_df['reading score'], color = 'pink')
plt.ylabel('Reading Score')
plt.xlabel('Math Score')
plt.title("Math and Reading Scores in Male and Female Students")
plt.legend((male, female),
           ('Male Students', 'Female Students'))


plt.tight_layout()
plt.show()
```

There is a strong positive correlation between math scores and reading scores in both male and female students. The female students have higher mean reading scores, and the male students have higher mean math scores.
